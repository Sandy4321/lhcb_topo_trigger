## Calculate thresholds on classifiers

```python
thresholds = dict()
for name, cl in estimators.items():
    prob = cl.predict_proba(ds_test_bck)
    thr, result = calculate_thresholds(ds_test_bck, prob, total_test_bck_events, rates=[2500., 4000.])
    thresholds[name] = thr
    print name, result
```

## Final efficiencies for each mode

```python
from collections import defaultdict
```

```python
# dict((name, rate): dict(mode: eff))
train_modes_eff = defaultdict(OrderedDict)
statistic = defaultdict(list)
for name, cl in estimators.items():
    for mode in sig_train_modes_names:
        statistic['mode'].append(mode)
        statistic['classifier'].append(name)
        statistic['name'].append('$' + Samples[str(mode)]['root'] + '$')
        sig_mode = ds_test_signal[ds_test_signal['mode'] == mode]
        sig_prediction = cl.predict_proba(sig_mode)[:, 1]
        sig_event_prediction = voting_for_event_svr(sig_mode['event_number'], sig_prediction)
        for rate, thr in thresholds[name].items():
            exist_sig = numpy.sum(sig_event_prediction >= thr)
            eff = exist_sig * 1. / total_test_signal_events[mode]
            statistic[rate].append(eff * 100.)
            train_modes_eff[(name, rate)]['$' + Samples[str(mode)]['root'] + '$\n' + str(mode)] = eff
```

```python
from rep.plotting import BarComparePlot
BarComparePlot(train_modes_eff).plot(new_plot=True, figsize=(18, 6), ylabel='efficiency')
legend(bbox_to_anchor=(0.5, 1.2), loc='upper center', ncol=3)
```

#### Generate eff metrics

```python
mode_metrics = dict()
for mode in sig_train_modes_names:
    mode_metrics[mode] = dict()
    for rate in [2500., 4000.]:
        mode_metrics[mode][rate] = generate_topo_metric(ds_test_bck, ds_test_signal[ds_test_signal['mode'] == mode], 
                                                        total_test_bck_events, total_test_signal_events[mode], rate)
```

```python
def generate_select_mode(mode):
    def select_mode(pd):
        result = numpy.zeros(len(pd), dtype=bool)
        result[numpy.array(pd['mode']) == mode] = True
        result[numpy.array(pd['signal']) == 0] = True
        return result
    return select_mode
```

```python
staged_eff_plots = []
for mode in sig_train_modes_names:
    eff_metrics = OrderedDict()
    eff_metrics['mode {}, rate {}, eff'.format(mode, 4000.)] = mode_metrics[mode][4000.]
    eff_metrics['mode {}, rate {}, eff'.format(mode, 2500.)] = mode_metrics[mode][2500.]
    select_mode = generate_select_mode(mode)
    staged_eff_plots.append(report.learning_curve(eff_metrics, mask=select_mode, steps=10))
```

```python
for elem in staged_eff_plots:
    elem.plot()
```

# Best model; all channels efficiencies

```python
efficiencies = defaultdict(OrderedDict)
result_statistic = defaultdict(list)
for mode in empty_events.keys():
    if mode in set(sig_train_modes_names) or mode == bck_train_mode_name:
        continue
    df = pandas.read_csv(os.path.join(folder , 'mod_{}.csv'.format(mode)), sep='\t')
    if len(df) <= 0:
        continue
    total_events = statistic_length(df)['Events'] + empty_events[mode]
    result_statistic['mode'].append(mode)
    result_statistic['name'].append('$' + Samples[str(mode)]['root'] + '$')
    result_statistic['total events'].append(total_events)
    df = df[numpy.array(df['pass_nbody']) == 1]
    passed_events = statistic_length(df)['Events']
    result_statistic['passed nbody events'].append(passed_events)
    result_statistic['eff of nbody passed events before classifier'].append(passed_events * 100. / total_events)
    prob = ef_add.predict_proba(df)
    for rate, thresh in thresholds['add features'].items():
        eff = final_eff_for_mode(df, prob, total_events, thresh)
        result_statistic[rate].append(eff * 100.)
        efficiencies[rate]['$' + Samples[str(mode)]['root'] + '$\n' + str(mode)] = eff
```

```python
from rep.plotting import BarComparePlot
BarComparePlot(efficiencies).plot(new_plot=True, figsize=(18, 6), ylabel='efficiency')
legend(bbox_to_anchor=(0.5, 1.2), loc='upper center', ncol=3)
```
