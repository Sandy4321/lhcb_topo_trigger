# Generate data

```python
%%time
with open("skim.log", "w") as l:
    threads = []
    for ch,val in Samples.iteritems():
        threads.append(subprocess.Popen([work_dir + "HLT/skim", str(ch), str(val["mother"]), str(len(val["children"])),
                                         work_dir + "Topo"],
                                        stdout=subprocess.PIPE, stderr=subprocess.PIPE))

    for thread in threads:
        t,e = thread.communicate()
        l.write(t)
        l.write("\n")
        l.write(e)
        l.write("\n")

with open("generate_hlt1.log", "w") as l:
    threads = []
    for ch,val in Samples.iteritems():
        threads.append(subprocess.Popen([work_dir + "HLT/generate_csv", str(ch), str(val["mother"]), str(len(val["children"])),
                                         work_dir + "Topo"],
                                        stdout=subprocess.PIPE, stderr=subprocess.PIPE))

    for thread in threads:
        t,e = thread.communicate()
        l.write(t)
        l.write("\n")
        l.write(e)
        l.write("\n")
```
