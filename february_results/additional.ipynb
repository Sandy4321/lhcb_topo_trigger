# Concat files

```python
import pandas
import numpy
import os
```

```python
CHUNKSIZE = 5000
def concat_files(folder, names, outfile):
    head = True
    mode = 'w'
    for name in names:
        pd_iterator = pandas.read_csv(os.path.join(folder, name), sep='\t', iterator=True, chunksize=CHUNKSIZE)
        for pd in pd_iterator:
            pd.to_csv(outfile, header=head, index=False, sep='\t', mode=mode)
            head = False
            mode = 'a'
```

# Metric definition

```python
import numpy
```

```python
from pprint import pprint
from collections import defaultdict, OrderedDict

def calculate_thresholds(test_bck, y_pred_bck, total_bck_events, rates, id_column='event_number'):
    required_size = dict([(rate, int(rate * total_bck_events / 1e6)) for rate in rates])
    bck_event_prediction = voting_for_event_svr(test_bck[id_column], y_pred_bck[:, 1])
    thresholds = dict()
    result = dict()
    for rate, req_size in required_size.items():
        threshold = numpy.percentile(bck_event_prediction, 100. * (1 - req_size * 1. / len(bck_event_prediction)))
        thresholds[rate] = threshold
        exist_bck = numpy.sum(bck_event_prediction > threshold)
        result[rate] = (threshold, exist_bck, 1. * exist_bck / total_bck_events)
    return thresholds, result

def voting_for_event_svr(ids, prediction):
    df = pandas.DataFrame({'prediction': numpy.array(prediction), 'id': numpy.array(ids)})
    return numpy.array(df.groupby('id')['prediction'].max())
```

```python
def compute_n_events_pass_threshold(event_number, predictions, threshold):
    return (numpy.bincount(event_number, weights=predictions > threshold) > 0).sum()
```

```python
def generate_topo_metric(test_bck, test_sig_mode, total_bck_events, 
                         total_sig_mode_events, rate, id_column='event_number'):
    def topo_metric(y_true, y_pred, sample_weight=None):
        threshold, _ = calculate_thresholds(test_bck, y_pred[y_true == 0], 
                                         total_bck_events, [rate], id_column=id_column)
        threshold = threshold[rate]
        exist_sig = compute_n_events_pass_threshold(numpy.array(test_sig_mode[id_column]), y_pred[y_true==1, 1], threshold)
        return  exist_sig * 1. / total_sig_mode_events
    return topo_metric
```

```python
def final_eff_for_mode(test, prediction, total_events, threshold, id_column='event_number'):
    event_prediction = voting_for_event_svr(numpy.array(test[id_column]), prediction[:, 1])
    exist = numpy.sum(event_prediction >= threshold)
    return exist * 1. / total_events
```

```python
def voting_for_event_svr_old(data, prediction, vote=numpy.max):
    data['prediction'] = numpy.array(prediction)
    event_result = []
    for key, group in data.groupby(['mode', 'event_number']):
        event_result.append(vote(numpy.array(group["prediction"])))
    return numpy.array(event_result)
```

```python
def generate_topo_metric_old(test_bck, test_sig_mode, total_bck_events, 
                         total_sig_mode_events, rate, vote=numpy.max):
    def topo_metric(y_true, y_pred, sample_weight=None):
        sig_event_prediction = voting_for_event_svr(test_sig_mode, y_pred[:, 1][y_true==1], vote)
        threshold, _ = calculate_thresholds(test_bck, y_pred[y_true == 0], 
                                         total_bck_events, [rate])
        threshold = threshold[rate]
        exist_sig = numpy.sum(sig_event_prediction > threshold)
        return  exist_sig * 1. / total_sig_mode_events
    return topo_metric
```

# Flatness

```python
from rep import calc_util, plotting

def calc_flatness(test, probs, signal_labels, bins=20, thresholds=None):
    def voting_svr(data, prediction, vote):
        data["modify_prediction"] = numpy.array(data['presel']) * numpy.array(prediction)
        data["btau"] = numpy.array(data["btau"])
        event_result = dict()
        for key, group in data.groupby("# event_id"):
            event_result[key] = (vote(numpy.array(group["modify_prediction"])),
                                 max(numpy.array(group['presel'])), max(numpy.array(group['btau'])))
        return event_result
    test_sig = test[signal_labels == 1]
    plot_all = []
    for key, prob in probs.iteritems():
        event_mark = pandas.DataFrame(voting_svr(test_sig, prob[signal_labels == 1], numpy.max)).transpose()
        event_mark.columns = ['p', 'presel', 'btau']
        class_prob = numpy.ndarray(shape=(len(event_mark), 2))
        class_prob[:, 1] = numpy.array(event_mark['p'])
        class_prob[:, 0] = 1 - class_prob[:, 1]
        eff_sig = calc_util.get_spectator_efficiency_correlation(class_prob, [1] * len(event_mark), 
                                                                 numpy.array(event_mark["btau"]),
                                                                 bins_number=bins, thresholds=thresholds)

        plot_fig = plotting.FunctionsPlot(eff_sig)
        plot_fig.xlabel = 'btau'
        plot_fig.ylabel = 'Efficiency'
        plot_fig.title = 'Signal flatness %s' %key 
        plot_all.append(plot_fig)

    return plotting.GridPlot(2, *plot_all)
```

```python
from rootpy.io import root_open
from rootpy.tree import Tree
import root_numpy

def add_prediction(filenames, treename, base_treename, classifiers, labels):
    import array

    ind_pr = 0
    normalization = 0
    ind_train_test_label = 0
    
    all_svr = 0
    
    for filename in filenames:#['D', 'U']:        
        len_svr = root_numpy.root2array(filename,
                                        base_treename, branches=['hlt2_svr_chi2'])['hlt2_svr_chi2']
        all_svr += sum([len(i) for i in len_svr])
        max_len = max([len(i) for i in len_svr])

        root_file = root_open(filename, "update")
        tree = root_file[base_treename]

        tree_new = Tree(treename)

        pred = len(classifiers) * [numpy.array([0], dtype=numpy.object)]
        for i in xrange(len(pred)):
            pred[i] = numpy.zeros(max_len, dtype=numpy.float32)
        
        tr_te = array.array('I', [0])
        count_svr = array.array('I', [0])
        new_br_svr = tree_new.Branch('n_svrs', count_svr, 'n_svrs/I')
        if labels is not None:
            new_br = tree_new.Branch('train_test_label', tr_te, 'train_test_label/I')
        newBranches = [tree_new.Branch(key, pred[ind], '%s[n_svrs]/F' % key) 
                      for key, ind in zip(classifiers.keys(), range(len(pred)))]
        
        
        for ind in xrange(len(tree)):
            tree.GetEntry(ind)
            count_svr[0] = len(len_svr[ind])
#             if normalization > 0:
#                 print 'a', count_svr[0]
#                 print labels[ind_train_test_label], ind + normalization, ind_train_test_label
            if count_svr[0] > 0:
                for num, (key, value) in enumerate(classifiers.iteritems()):
                    pred[num][0:count_svr[0]] = value[ind_pr:ind_pr+count_svr[0]]
                ind_pr += count_svr[0]
            
                tr_te[0] = 0
                if labels is not None and labels[ind_train_test_label][1] == ind + normalization:
    #                 if normalization > 0:
    #                     print labels[ind_train_test_label],
                    assert labels[ind_train_test_label][3] == count_svr[0], ind
                    tr_te[0] = labels[ind_train_test_label][0]
    #                 print count[0] 
                    ind_train_test_label += 1
#             new_br.Fill()
#             new_br_svr.Fill()
#             for br in newBranches:
#                 br.Fill()
            tree_new.Fill()
        tree_new.write()
        normalization += len(tree)
        root_file.close()
    if labels is not None:
        assert  len(labels) == ind_train_test_label, 'ERROR'
    assert  ind_pr == all_svr, 'ERROR'
```
