# Metric definition

```python
from pprint import pprint
from collections import defaultdict, OrderedDict
import pandas

empty_events = {11114001: 221, 
                11124001: 137,
                11296013: 33,
                11874042: 219, 
                12103035: 533,
                12165106: 463,
                12265042: 79,
                30000000: 107352}
    
def metric_generate(test, vote=numpy.max, metric_plot=False, channel_vote=numpy.min):
    bkg = test[test['signal'] == 0]
    total_size_bkg = len(numpy.unique(bkg['# event_id'])) + int(empty_events[30000000] / 2)
    
    # corresponds to rate (Please CHECK!!!!)
    required_size = int(2500. / 1e6 * total_size_bkg)
    print required_size

    def metric_additional(y_true, y_pred, sample_weight=None):
        event_mark = pandas.DataFrame(voting_svr(test, y_pred, vote)).transpose()
        event_mark.columns = ['p', 'channel', 'presel', 'bdt_passed', 'bdt_passed_modified']
        channels = set(map(int, event_mark['channel']))
        bkg_pred = array(event_mark[event_mark['channel'] == 30000000]['p'])
        print required_size
        threshold = numpy.sort(bkg_pred)[-required_size - 1]
        print threshold
        print len(numpy.where(bkg_pred > threshold)[0])
        result_eff = dict()
        efficiencies = []
        for channel in channels:
            result_eff[channel] = dict()
            result_eff[channel]["total"] = numpy.sum(event_mark['channel'] == channel)
            result_eff[channel]["total_empty"] = numpy.sum(event_mark['channel'] == channel) + int(empty_events[channel] / 2)
            result_eff[channel]["passed"] = numpy.sum((event_mark['channel'] == channel) & (event_mark['p'] > threshold))
            result_eff[channel]["p"] = 1.0 * result_eff[channel]["passed"] / result_eff[channel]["total"]
            result_eff[channel]["p_empty"] = 1.0 * result_eff[channel]["passed"] / result_eff[channel]["total_empty"]
            result_eff[channel]["bdt_passed"] = 1.* numpy.sum((event_mark['channel'] == channel) & (event_mark['bdt_passed'] > 0)) / \
                                                result_eff[channel]["total_empty"]
            result_eff[channel]["bdt_passed_modified"] = 1. * numpy.sum((event_mark['channel'] == channel) & 
                                                                   (event_mark['bdt_passed_modified'] > 0)) / \
                                                         result_eff[channel]["total_empty"]
            result_eff[channel]["eff_preselected"] = \
                1.0 *  numpy.sum((event_mark['channel'] == channel) & (event_mark['presel'] == 1)) / result_eff[channel]["total_empty"]
            if channel != 30000000:
                efficiencies.append(result_eff[channel]["p_empty"])
        return efficiencies, result_eff
    
    def metric(y_true, y_pred, sample_weight=None):
        eff, data = metric_additional(y_true, y_pred, sample_weight=sample_weight)
        if not metric_plot:
            return channel_vote(eff)
        else:
            return data
    return metric
    
def voting_svr(data, prediction, vote):
    data["bdt_passed"] = bdt_Mike(data)
    data["bdt_passed_modified"] = bdt_Mike_modified(data)
    data["modify_prediction"] = numpy.array(data['presel']) * numpy.array(prediction)
    event_result = dict()
    for key, group in data.groupby("# event_id"):
        event_result[key] = (vote(numpy.array(group["modify_prediction"])), group["channel_label"].iloc[0],
                             max(numpy.array(group['presel'])), vote(group["bdt_passed"]), 
                             vote(group["bdt_passed_modified"]))
    return event_result

def bdt_Mike(data):
    data["bdt_cut"] = 0.4
    data["bdt_cut"][data["n"] == 4] = 0.3
    data["bdt_cut"][data["nmu"] > 0] = 0.1
    passed = numpy.ones(len(data))
    passed[array(data["bdt"]) < array(data["bdt_cut"])] = 0
    passed[data["fdchi2"].values < 100] = 0
    passed[data["sumpt"].values < 4000] = 0
    passed[(data["n"].values > 2) * (data["sumpt"].values < 4500)] = 0
    passed[data["minpt"].values < 500] = 0
    passed[data["nhlt1"].values == 0] = 0
    return passed

def bdt_Mike_modified(data):
    data["bdt_cut_m"] = 0.3
    data["bdt_cut_m"][data["nmu"] > 0] = 0.1
    passed = numpy.ones(len(data))
    passed[array(data["bdt"]) < array(data["bdt_cut_m"])] = 0
    passed[data["fdchi2"].values < 32] = 0
    return passed
```

# Plot definition

```python
def plot_eff(data_dict, p_label='p', total_label='total', alpha=0.4, move=3, bar_width=1, length=8):
    for key, value in data_dict.iteritems():
        data = value['data']
        p_label_ = value['p_label'] if 'p_label' in value else p_label
        total_label_ = value['total_label'] if 'total_label' in value else total_label
        alpha_ = value['alpha'] if 'alpha' in value else alpha
        move_ = value['move'] if 'move' in value else move
        color = value['color']
        
        error_config = {'ecolor': '0.3'}
        modes = sort(list(set(data.keys())))# - {30000000}))
        if length is None:
            length = len(modes)
        
        index = np.arange(len(modes))
        
        plt.bar(length * index + move_, [1 - data[mode][p_label_] for mode in modes], bar_width,
                       alpha=alpha_, color=color, 
                       yerr=[sqrt(data[mode][p_label_] * (1 - data[mode][p_label_]) / data[mode][total_label_])
                                                   for mode in modes], error_kw=error_config,
                       label=key)
        plt.xticks(length * index + move_ * bar_width, modes)
```

```python
class PlotMetric:
    def __init__(self, metrics, y_true, prob, sample_weight=None):
        self.result = {}
        for key, func in metrics.iteritems():
            self.result[key] = func(y_true, prob, sample_weight=sample_weight)
    def plot(self, color='b', size=20, label='classifier', x_push=0, y_push=0):
        x = range(len(self.result.keys()))
        y = self.result.values()
        plt.scatter(x, y, color=color, s=size, label=label)
        for x_i, y_i in zip(x, y):
            plt.annotate('%1.5f' % y_i, xy = (x_i, y_i), 
                         xytext = (x_i + 0.05 + x_push, y_i + 0.05 + y_push), size='medium', color=color)
        plt.xticks(range(len(self.result.keys())), self.result.keys())
```

```python
def memory_usage():
    """Memory usage of the current process in bytes. Created for notebooks.
    This will only work on systems with a /proc file system (like Linux)."""
    result = {'peak': 0, 'rss': 0}
    with open('/proc/self/status') as status:
        for line in status:
            parts = line.split()
            key = parts[0][2:-1].lower()
            if key in result:
                result[key] = "{:,} kB".format(int(parts[1]))
    return result
```

# Flatness

```python
from cern_utils import calc_util, plotting

def calc_flatness(test, probs, signal_labels, bins=20, thresholds=None):
    def voting_svr(data, prediction, vote):
        data["modify_prediction"] = numpy.array(data['presel']) * numpy.array(prediction)
        data["btau"] = numpy.array(data["btau"])
        event_result = dict()
        for key, group in data.groupby("# event_id"):
            event_result[key] = (vote(numpy.array(group["modify_prediction"])),
                                 max(numpy.array(group['presel'])), max(numpy.array(group['btau'])))
        return event_result
    test_sig = test[signal_labels == 1]
    plot_all = []
    for key, prob in probs.iteritems():
        event_mark = pandas.DataFrame(voting_svr(test_sig, prob[signal_labels == 1], numpy.max)).transpose()
        event_mark.columns = ['p', 'presel', 'btau']
        class_prob = numpy.ndarray(shape=(len(event_mark), 2))
        class_prob[:, 1] = numpy.array(event_mark['p'])
        class_prob[:, 0] = 1 - class_prob[:, 1]
        eff_sig = calc_util.get_spectator_efficiency_correlation(class_prob, [1] * len(event_mark), 
                                                                 numpy.array(event_mark["btau"]),
                                                                 bins_number=bins, thresholds=thresholds)

        plot_fig = plotting.FunctionsPlot(eff_sig)
        plot_fig.xlabel = 'btau'
        plot_fig.ylabel = 'Efficiency'
        plot_fig.title = 'Signal flatness %s' %key 
        plot_all.append(plot_fig)

    return plotting.GridPlot(2, *plot_all)
```

```python
from rootpy.io import root_open
from rootpy.tree import Tree
import root_numpy

def add_prediction(filenames, treename, base_treename, classifiers, labels):
    import array

    ind_pr = 0
    normalization = 0
    ind_train_test_label = 0
    
    all_svr = 0
    
    for filename in filenames:#['D', 'U']:        
        len_svr = root_numpy.root2array(filename,
                                        base_treename, branches=['hlt2_svr_chi2'])['hlt2_svr_chi2']
        all_svr += sum([len(i) for i in len_svr])
        max_len = max([len(i) for i in len_svr])

        root_file = root_open(filename, "update")
        tree = root_file[base_treename]

        tree_new = Tree(treename)

        pred = len(classifiers) * [numpy.array([0], dtype=numpy.object)]
        for i in xrange(len(pred)):
            pred[i] = numpy.zeros(max_len, dtype=numpy.float32)
        
        tr_te = array.array('I', [0])
        count_svr = array.array('I', [0])
        new_br_svr = tree_new.Branch('n_svrs', count_svr, 'n_svrs/I')
        if labels is not None:
            new_br = tree_new.Branch('train_test_label', tr_te, 'train_test_label/I')
        newBranches = [tree_new.Branch(key, pred[ind], '%s[n_svrs]/F' % key) 
                      for key, ind in zip(classifiers.keys(), range(len(pred)))]
        
        
        for ind in xrange(len(tree)):
            tree.GetEntry(ind)
            count_svr[0] = len(len_svr[ind])
#             if normalization > 0:
#                 print 'a', count_svr[0]
#                 print labels[ind_train_test_label], ind + normalization, ind_train_test_label
            if count_svr[0] > 0:
                for num, (key, value) in enumerate(classifiers.iteritems()):
                    pred[num][0:count_svr[0]] = value[ind_pr:ind_pr+count_svr[0]]
                ind_pr += count_svr[0]
            
                tr_te[0] = 0
                if labels is not None and labels[ind_train_test_label][1] == ind + normalization:
    #                 if normalization > 0:
    #                     print labels[ind_train_test_label],
                    assert labels[ind_train_test_label][3] == count_svr[0], ind
                    tr_te[0] = labels[ind_train_test_label][0]
    #                 print count[0] 
                    ind_train_test_label += 1
#             new_br.Fill()
#             new_br_svr.Fill()
#             for br in newBranches:
#                 br.Fill()
            tree_new.Fill()
        tree_new.write()
        normalization += len(tree)
        root_file.close()
    if labels is not None:
        assert  len(labels) == ind_train_test_label, 'ERROR'
    assert  ind_pr == all_svr, 'ERROR'
```
