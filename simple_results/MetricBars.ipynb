## Without empty events

###### We use the same thresholds for all classifiers. In fact, we just moved from "p = 1 - N_passed/N_total" to "p = 1 - N_passed/(N_total - N_empty)"

```python
fig, ax = plt.subplots(figsize=(13,7))

bar_width = 1
modes = sorted(matrixnet_quality.keys())
modes.remove(30000000)
index = np.arange(len(modes))
opacity = 0.4
error_config = {'ecolor': '0.3'}
rec1 = plt.bar(7 * index + 3, [1 - 1. * matrixnet_quality[mode]['passed'] / (matrixnet_quality[mode]['total'] - int(empty_events[mode] / 2)) for mode in modes], bar_width,
        alpha=1,
        color='g',
        yerr=[matrixnet_quality[mode]['error'] for mode in modes],
        error_kw=error_config,
        label='Matrixnet')

rec2 = plt.bar(7 * index + 4, [1 - 1. * adaboost_quality[mode]['passed'] / (adaboost_quality[mode]['total'] - int(empty_events[mode] / 2)) for mode in modes], bar_width,
        alpha=opacity,
        color='b',
        yerr=[adaboost_quality[mode]['error'] for mode in modes],
        error_kw=error_config,
        label='Adaboost')

rec3 = plt.bar(7 * index + 5, [1 - 1. * rf_quality[mode]['passed'] / (rf_quality[mode]['total'] - int(empty_events[mode] / 2)) for mode in modes], bar_width,
        alpha=opacity,
        color='r',
        yerr=[rf_quality[mode]['error'] for mode in modes],
        error_kw=error_config,
        label='RandomForest')

bdt_passed_modified_p = {mode:
    1 - (1. * matrixnet_quality[mode]['bdt_passed_modified']
         * matrixnet_quality[mode]['total']
         / (matrixnet_quality[mode]['total'] - int(empty_events[mode] / 2))) 
    for mode in modes
}
bdt_passed_modified_err = [
    sqrt(
        bdt_passed_modified_p[mode] * (1 - bdt_passed_modified_p[mode]) 
        / (matrixnet_quality[mode]['total'] - int(empty_events[mode] / 2)))
    for mode in modes
]
rec4 = plt.bar(
        7 * index + 6,
        [bdt_passed_modified_p[mode] for mode in modes], bar_width,
        alpha=opacity,
        color='pink',
        yerr=bdt_passed_modified_err,
        error_kw=error_config,
        label='BDT modified')

bdt_passed_p = {mode:
    1 - (1. * matrixnet_quality[mode]['bdt_passed']
         * matrixnet_quality[mode]['total']
         / (matrixnet_quality[mode]['total'] - int(empty_events[mode] / 2))) 
    for mode in modes
}
bdt_passed_err = [
    sqrt(
        bdt_passed_p[mode] * (1 - bdt_passed_p[mode]) 
        / (matrixnet_quality[mode]['total'] - int(empty_events[mode] / 2)))
    for mode in modes
]
rec5 = plt.bar(7 * index + 7, [bdt_passed_p[mode] for mode in modes], bar_width,
        alpha=opacity,
        color='cyan',
        yerr=bdt_passed_err,
        error_kw=error_config,
        label='BDT')

plt.xlabel('Modes')
plt.ylabel('Inefficiency')
plt.title('Classifier Inefficiency')
plt.xticks(7 * index + 6 * bar_width, modes)
plt.legend(bbox_to_anchor=(0.17, 0.99))

plt.tight_layout()
```
