```python
def get_channel_labels(ds):
    channel_label = []
    for element in ds['# event_id']:
        channel_label.append(int(element.split('_')[0]))
    return channel_label
```

## Channel classifiers

```python
channel_mns = []
for channel in channels_all[:-1]:
    ch = channel[4:-4]
    mn_t = ef_classifier.ClassifierEF(directory=cl_folder + 'channel_' + ch, formula_name='topo_mn_train',
                                      dataset_name='topo_mn_channel_' + ch,
                                      url_base='test02cern-i.vs.os.yandex.net')
    mn_t.set_params(features=variables_base, iterations=3000, sync=False)
    mn_t.fit(sig_data_stage.select('channel_label==%s' %ch), bck_data_stage)
    channel_mns.append(mn_t)
```

```python
for cl, name in zip(channel_mns, channels_all[:-1]):
    importance = cl.get_feature_importance()
    importance.sort(['effect'], ascending=False)[['effect']].plot(figsize=(13,3), kind='bar', title=name[4:-4])
```

```python
train_cols_s = {}
train_cols_b = {}

for cl, ch in zip(channel_mns, channels_all[:-1]):
    train_cols_s['mn_' + ch[4:-4]] = cl.predict(sig_data_train)
    train_cols_b['mn_' + ch[4:-4]] = cl.predict(bck_data_train)
for cl, k in zip([mn_body2, mn_body3, mn_body4], [2, 3, 4]):
    train_cols_s['mn_body_%d' % k] = cl.predict(sig_data_train)
    train_cols_b['mn_body_%d' % k] = cl.predict(bck_data_train)
```

```python
sig_data_train = sig_data_train.add_columns(train_cols_s)
bck_data_train = bck_data_train.add_columns(train_cols_b)
```
